```python
# Librerías
import pandas as pd
import torch
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from sklearn.metrics import roc_auc_score, classification_report
from sklearn.preprocessing import LabelBinarizer

# Cargar dataset
from google.colab import drive
drive.mount('/content/drive')
ruta = '/content/data_test_fold2 - anorexia(in).csv'

df = pd.read_csv(ruta)

clase = df.iloc[:, 3]
clases = clase[:250] #1 --> Anorexia, 0 --> No anorexia

# Cargar tokenizer y modelo
model_name = "joeddav/xlm-roberta-large-xnli"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)

# Etiquetas del modelo
labels = ['contradiction', 'neutral', 'entailment']

# Verifica si tiene una GPU compatible con CUDA
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

# Hipótesis para detección de anorexia
hipotesis = "La persona muestra signos de anorexia/bulimia."

# Función de inferencia
def inferencia_xlm_roberta(premisa, hipotesis):
    inputs = tokenizer.encode_plus(
        premisa,
        hipotesis,
        return_tensors='pt',
        truncation=True,
        padding=True,
        max_length=512
    ).to(device)

    with torch.no_grad():
        logits = model(**inputs).logits
        probs = torch.softmax(logits, dim=1).cpu().numpy()[0]

    return dict(zip(labels, probs))

# Aplicar al dataset
entailments, neutrals, contradictions = [], [], []

# Obtener inferencias del dataset X(Resultado limpio)
for texto in X:
    # Convertir la lista de palabras en una cadena de texto
    texto_str = " ".join(texto)
    scores = inferencia_xlm_roberta(texto_str, hipotesis)
    entailments.append(scores['entailment'])
    neutrals.append(scores['neutral'])
    contradictions.append(scores['contradiction'])


# Guardar resultados
df['entailment'] = entailments
df['neutral'] = neutrals
df['contradiction'] = contradictions

# Clasificación binaria
df['pred_clase'] = df['entailment'].apply(lambda x: 'anorexia' if x > 0.5 else 'control')

# Evaluación
lb = LabelBinarizer()
y_true = lb.fit_transform(clases)
y_score = df['entailment']

auc = roc_auc_score(y_true, 1- y_score)
print(f"\n AUC-ROC: {auc:.4f}")
print("\n Classification Report:")
print(classification_report(clases, df['pred_clase']))

# Ver cada tweet con su predicción (las 10 primeras)
# for index, row in df.head(10).iterrows():
#     print(f"Tweet: {row['tweet_text']}")
#     print(f"Predicción: {row['pred_clase']}, Clase: {row['class']}")
#     print(f"Probabilidad de Anorexia: {row['entailment']:.4f}\n")
```
